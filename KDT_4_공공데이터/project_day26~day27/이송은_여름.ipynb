```python
def Dragonflies(city,location,c_color):
  file=pd.read_csv('/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/(4) 공공데이터/프로젝트/OBS_계절관측.csv', encoding='euc_kr')
  file_df=pd.DataFrame(file)
  file_df_Dragonflies=file_df.loc[:,['지점','년도','잠자리','잠자리.1','잠자리.2','잠자리.3','잠자리.4','잠자리.5','잠자리.6','잠자리.7']]
  file_df_Dragonflies.replace(['관측 안됨', ' ','―','결측'], np.NAN, inplace=True)
  file_df_Dragonflies.drop(columns=['잠자리.2','잠자리.3','잠자리.4','잠자리.5','잠자리.6','잠자리.7'], inplace=True)
  file_df_Dragonflies.dropna(inplace=True)
  file_df_Dragonflies_city=file_df_Dragonflies[file_df_Dragonflies['지점']==city].copy()
  file_df_Dragonflies_city['년도']=file_df_Dragonflies_city['년도'].astype(int)
  file_df_Dragonflies_city['평비']=file_df_Dragonflies_city['잠자리.1'].str.replace('일','').astype(int)

  plt.figure(figsize=(10,3))
  plt.plot(file_df_Dragonflies_city['년도'],file_df_Dragonflies_city['평비'],color=c_color)
  plt.xticks(file_df_Dragonflies_city['년도'],list(range(2005,2025)))
  plt.grid(True,linewidth=0.5)
  plt.title(f'{city}({location})에서 년도별 잠자리 평균대비 추세')
  plt.xlabel('년도')
  plt.ylabel('평균대비')
```

```python
def Cicada(city,location,c_color):
  file=pd.read_csv('/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/(4) 공공데이터/프로젝트/OBS_계절관측.csv', encoding='euc_kr')
  file_df=pd.DataFrame(file)
  file_df_cicada=file_df.loc[:,['지점','년도','매미','매미.1','매미.2','매미.3','매미.4','매미.5','매미.6','매미.7']]
  file_df_cicada.replace(['관측 안됨', ' ','―','결측'], np.NAN, inplace=True)
  file_df_cicada.drop(columns=['매미','매미.1','매미.2','매미.3','매미.6','매미.7'], inplace=True)
  file_df_cicada.dropna(inplace=True)
  file_df_cicada_city=file_df_cicada[file_df_cicada['지점']==city].copy()
  file_df_cicada_city['년도']=file_df_cicada_city['년도'].astype(int)
  file_df_cicada_city['평비']=file_df_cicada_city['매미.5'].str.replace('일','').astype(int)

  plt.figure(figsize=(10,3))
  plt.plot(file_df_cicada_city['년도'],file_df_cicada_city['평비'],color=c_color)
  plt.xticks(file_df_cicada_city['년도'],list(range(2005,2025)))
  plt.grid(True,linewidth=0.5)
  plt.title(f'{city}({location})에서 년도별 매미 평균대비 추세')
  plt.xlabel('년도')
  plt.ylabel('평균대비')
```
