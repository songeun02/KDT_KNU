```python
file_s=file[['survived','sex','age','adult_male','deck','embark_town','alive']]
file_s
```

```python
file_s_c=file_s[(file_s['alive']=='yes') & (file_s['sex']=='male' )].copy()
file_s_c
```

```python
file_s_c.isna().sum()
```

```python
file_s_c.fillna({'deck':'unknown'},inplace=True)
```

```python
file_s_c
```

```python
file_s_c.isna().sum()
```

특정 컬럼인 'age'만 NaN 여부 검사 진행

```python
file_s_c=file_s_c.dropna(subset=['age']).copy()
```
