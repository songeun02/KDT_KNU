### 데이터의 분위수 구하는 quantile()

```python
quantile=[0,0.2,0.4,0.6,0.8,1]

for idx in quantile:
  q=df['Total_Review'].quantile(idx,interpolation='lower')
  print(f'quantile({idx}) is {q}')
```

```python
for idx in quantile:
  q=df['Total_Review'].quantile(idx, interpolation='nearest')
  print(f'quantile({idx}) is {q}')
```

### 그룹별 연산을 위한 groupby()

```python
df.groupby('sex')['survived'].mean()
```

```python
df.groupby(['sex','pclass'])['survived'].mean()
```

```python
df.groupby(['sex','pclass'])['survived'].agg(['mean','count'])
```

```python
df.groupby(['sex','pclass'])[['survived','age']].agg({'survived':'mean','age':'max'})
```

```python
def get_iqr(data):
  _3rd=data.quantile(0.75)
  _1rd=data.quantile(0.25)
  return (np.abs(_3rd-_1rd)*1.5)
```

```python
pd.DataFrame(df.groupby(['sex','pclass'])['age'])
```

```python
df.groupby(['sex','pclass'])['age'].apply(get_iqr)
```
