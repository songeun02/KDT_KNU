```python
size = int(input("홀수차 배열의 크기를 입력하세요 : "))
print(f"Magic Square ({size}x{size})")
size_array=[[0]*size for i in range(size)]



if size%2==1:
  row=0
  col=size//2
  size_array[row][col]=1
  for i in range(2,size**2+1):
    row-=1
    col+=1
    if row<0 or row>size-1:
      row=size-1


    if col<0 or col>size-1:
      col=0

    if size_array[row][col]!=0:
      row+=2
      col-=1

      if row<0 or row>size-1:
        row-=size

      if col<0 or col>size-1:
        col+=size

    size_array[row][col]=i


  for i in range(size):
    for j in range(size):
      print(f'{size_array[i][j]:>2}', end='')
    print()



else:
  print("짝수를 입력하였습니다. 다시 입력하세요.")
```
