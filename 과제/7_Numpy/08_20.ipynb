```python
# 질의 함수
# 넘파이 함수 choose는 인덱스로 검색

x = np.arange(9).reshape(3,3)
x
```

```python
np.choose(0,x) # np.choose(choices, arrays)
```

```python
np.choose([0],x)
```

```python
np.choose(tuple([0,1,2]), x) # x의 각 행에서 해당 위치 값 선택
```

```python
np.choose(tuple([2,1,0]),x)

# 첫 번째 열에서는 세번째 행 -> [2],[0]
# 두 번째 열에서는 두번째 행 -> [1],[1]
# 세 번째 열에서는 첫번째 행 -> [0],[2]
```

```python
# 넘파이 where은 조건식 판단을 검색에 넣어서 조회 가능 -> select, extract

np.where(x>4) # 조건 만족하는 인덱스 찾음

# 2차원에서는 앞의 array는 axis = 0, 뒤의 array는 axis = 1에 해당
# -> [1,2],[2,0],[2,1],[2,2]
```

```python
a = np.select(x>5, x*2) # 앞의 조건에서 만족한 것들을 대상으로 적용할 값
a
```

```python
np.extract(x>4, x)
```
