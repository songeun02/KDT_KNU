```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
import csv
```

```python
with open('/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/data/공공데이터/07_29/subwaytime.csv', encoding='utf-8-sig') as f:
  data=csv.reader(f)
  next(data)
  next(data)
  max_person=0
  max_station=['']*7
  max_person=[0]*7
  station=['1호선','2호선','3호선','4호선','5호선','6호선','7호선']

  for row in data:
    row[4:]=map(int,row[4:])
    seven_nine=row[11]+row[13]
    for i in range(len(station)):
      if row[1]==station[i]:
        if seven_nine>max_person[i]:
          max_person[i]=seven_nine
          max_station[i]=row[3]


  for i in range(len(station)):
    print(f'출근 시간대 {station[i]} 최대 하차역 : {max_station[i]}역, 하차인원 : {max_person[i]:,}명')


name=[station[i]+ " " + max_station[i] for i in range(len(station))]


plt.bar(name, max_person)
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역')
plt.xticks(rotation=80)
plt.show()
```
