```python
import random
```

```python
def score_count(sticks):
  score=0
  score_yut=" "
  if sticks==[1,1,1,1]:
    score=5
    score_yut="모"
  elif sticks==[0,0,0,0]:
    score=4
    score_yut="윷"
  elif sticks==[1,0,0,0]:
    score=3
    score_yut="걸"
  elif sticks==[1,1,0,0]:
    score=2
    score_yut="개"
  elif sticks==[1,1,1,0]:
    score=1
    score_yut="도"
  return score, score_yut
```

```python
def play_yut():

  h_count=0
  n_count=0

  order=0

  sticks=[0,0,0,0]

  high_20=False

  while 1:
    while order==0:
      for i in range(4):
        sticks[i]=random.randint(0,1)
      score_h, score_yut_h=score_count(sticks)
      h_count+=score_h
      if score_yut_h in ["모","윷","걸","개","도"]:
        print(f'흥부 {sticks} : {score_yut_h} ({score_h}점)/(총 {h_count}점) --->')
      else:
        continue
      if score_yut_h in ["모","윷"]:
        if h_count>=20:
          high_20=True
        else:
          order=0
          continue
      if h_count>=20:
        high_20=True
        print('-'*100)
        print(f'흥부 승리 => 흥부 : {h_count}, 놀부 : {n_count}')
        print('-'*100)
        break
      order=1
    while order==1:
      for i in range(4):
        sticks[i]=random.randint(0,1)
      score_n, score_yut_n=score_count(sticks)
      n_count+=score_n
      if score_yut_n in ["모","윷","걸","개","도"]:
        print("{" ": >80}".format(f'<--- 놀부 {sticks} : {score_yut_n} ({score_n}점)/(총 {n_count}점)'))
      else:
        continue
      if score_yut_n in ["모","윷"]:
        if n_count>=20:
          high_20=True
        else:
          order=1
          continue
      if n_count>=20:
        high_20=True
        print('-'*100)
        print(f'놀부 승리 => 흥부 : {h_count}, 놀부 : {n_count}')
        print('-'*100)

        break
      order=0

    if high_20==True:
      break



```
