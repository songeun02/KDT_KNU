```python
import csv
import matplotlib.pyplot as plt
import koreanize_matplotlib
import re
```

```python
f=open('/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/data/공공데이터/07_30/gender.csv', encoding='euc_kr')
data=csv.reader(f)

male_list=[]
female_list=[]
city_list=[]
values=[]


for row in data:
  if ('대구광역시' in row[0]) :
    male_list.append(int(row[104].replace(',','')))
    female_list.append(int(row[207].replace(',','')))
    city_list.append(re.split('[()]', row[0]))

for i in range(len(city_list)):
  print(f'{city_list[i][0]} : (남:{male_list[i]:,} 여:{female_list[i]:,})')
  values.append([male_list[i], female_list[i]])

fig=plt.figure(figsize=(7,10))
axes=fig.subplots(5,2)

index=0
for i in range(5):
  for j in range(2):
    axes[i,j].pie(values[index], autopct='%.1f%%', startangle=90, textprops={'fontsize':8}, labels=['남성','여성'])
    axes[i,j].set_title(city_list[index][0], fontsize=10)

    index+=1

fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=15)
plt.show()

```
