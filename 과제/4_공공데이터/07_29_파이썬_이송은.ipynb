```python
def manu():
  print('-'*70)
  print("1. 전체 데이터 출력")
  print("2. 수도 이름 오름차순 출력")
  print("3. 모든 도시의 인구수 내림차순 출력")
  print("4. 특정 도시의 정보 출력")
  print("5. 대륙별 인구수 계산 및 출력")
  print("6. 프로그램 종료 ")
  print('-'*70)

country_dict={'Seoul':['South Korea','Asia',9655000],
              'Tokyo':['Japan','Asia',14110000],
              'Beijing':['China','Asia',21540000],
              'London':['United Kingdom','Europe',14800000],
              'Berlin':['Germany','Europe',3426000],
              'Mexico City':['Mexico','America',21200000]}

area=['Asia','Europe','America']

while 1:
  manu()
  check_num=int(input("메뉴를 입력하세요 : "))
  if check_num==1:
    index=0
    for key, value in  country_dict.items():
      print(f'[{index+1}] {key} : {value}')
      index+=1
  if check_num==2:
    index=0
    sorted_country_dict=dict(sorted(country_dict.items()))
    for key, value in  sorted_country_dict.items():
      print(f'[{index+1}] {key:<12} : {value[0]:<15} {value[1]:<10} {value[2]:>10,}')
      index+=1
  if check_num==3:
    index=0
    coutry_dict_list=sorted(country_dict.items(), key=lambda x:x[1][2], reverse=True)
    for i in range(6):
      print(f"[{index+1}] {coutry_dict_list[i][0]:<12} : {coutry_dict_list[i][1][2]:>12,}")
      index+=1

  if check_num==4:
    search=input("출력할 도시 이름을 입력하세요 : ")
    if search in country_dict.keys():
      print(f'도시 : {search}')
      print(f'국가 : {country_dict[search][0]}, 대륙 : {country_dict[search][1]}, 인구수 : {country_dict[search][2]:,}')
    else:
      print(f'도시이름 : {search}은 key에 없습니다.')

  if check_num==5:
    input_area=input("대륙 이름을 입력하세요(Asia, Europe, America) : ")
    sum_area=0
    for key, value in  country_dict.items():
      if value[1]==input_area:
        print(f'{value[0]} : {value[2]:,}')
        sum_area+=value[2]
    print(f'{input_area} 전체 인구 수 : {sum_area:,}')

  if check_num==6:
    print("프로그램을 종료합니다")
    break
```
