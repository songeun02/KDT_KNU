```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

def scraping_use_find(html):

  print('[find 함수 사용]')

  find_tom_c = html.find_all('div',{'class':'tombstone-container'})

  print('총 tomestone-container 검색 개수 : ', len(find_tom_c))

  print('*'*80)

  for weather_find in find_tom_c:

    print(f"[Period]:{weather_find.find('p',{'class':'period-name'}).text}")
    print(f"[Short desc] : {weather_find.find('p',{'class':'short-desc'}).text}")
    print(f"[Temperature] : {weather_find.find('p',{'class':'temp'}).text}")
    print(f"[Image desc] : {weather_find.find('img')['title']}")
    print('-'*80)


def scraping_use_select(html):

  print('[select 함수 사용]')

  select_tom_c = html.select('div.tombstone-container')

  print('총 tomestone-container 검색 개수 : ', len(select_tom_c))

  print('*'*80)

  for weather_find in select_tom_c:

    print(f"[Period]:{weather_find.select_one('p.period-name').text}")
    print(f"[Short desc] : {weather_find.select_one('p.short-desc').text}")
    print(f"[Temperature] : {weather_find.select_one('p.temp').text}")
    print(f"[Image desc] : {weather_find.select_one('img')['title']}")
    print('-'*80)


def main():
  page = urlopen(' https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Yst5ji9yxTY')
  html = BeautifulSoup(page.read(), 'html.parser')

  print('National Weather Service Scraping')
  print('*'*80)

  scraping_use_find(html)
  scraping_use_select(html)

main()
```
