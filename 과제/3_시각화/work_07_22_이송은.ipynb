```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t_df_s_c_df_f=t_df_s_c_df[t_df_s_c_df['sex']=='female']
t_df_s_c_df_f=t_df_s_c_df_f.reindex([1,0,2,3,4,5,6])
t_df_s_c_df_f
```

```python
t_df_s_c_df_m=t_df_s_c_df[t_df_s_c_df['sex']=='male']
t_df_s_c_df_m=t_df_s_c_df_m.reindex([8,7,9,10,11,12,13,14])
t_df_s_c_df_m
```

```python
plt.plot(t_df_s_c_df_f['age_group'],t_df_s_c_df_f['count'],label='Female')
plt.plot(t_df_s_c_df_m['age_group'],t_df_s_c_df_m['count'],label='male')

plt.xlabel("나이대")
plt.ylabel("나이대별 count ")
plt.legend()

plt.show()
```

```python
p_df_all=p_df[p_df['전출지별']=="전국"].copy()
p_df_all
```

```python
p_df_all.replace('-',0,inplace=True)
p_df_all.drop(1,inplace=True)
p_df_all.reset_index(inplace=True)
p_df_all.drop('index',inplace=True,axis=1)
p_df_all
```

```python
p_df_all.set_index(['전출지별','전입지별'], inplace=True)
```

```python
p_df_all.info()
```

```python
p_df_all.max()
```

```python
p_df_all.idxmax()
```

```python
p_df_all.columns
```

```python
p_df_all_sub=pd.DataFrame()
```

```python
for i in range(1970,2017):
  p_df_all_sub[f'{i+1}-{i}']=abs(p_df_all[f'{i+1}']-p_df_all[f'{i}']) # 인원 이동 변동을 보는 것이므로 절대값 사용
```

```python
p_df_all_sub
```

```python
p_df_all_sub.info()
```

```python
p_df_all_sub.max()
```

```python
p_df_all_sub.idxmax()
```

```python
p_df_all_sub
```

```python
p_df_all_sub_index=pd.DataFrame(p_df_all_sub.index.to_list())
p_df_all_sub_index
```

```python
sub_list=p_df_all_sub.index.to_list()
```

```python
plt.figure(figsize=[20,5])

for i in range(len(sub_list)):
  list_city=p_df_all_sub.index.to_list()
  list_index_i=p_df_all_sub_index[1]
  plt.plot(p_df_all_sub.loc[list_city[i],:], label=list_index_i[i])
  plt.xticks(rotation=45)

plt.legend(loc='center right')
plt.show()
```
