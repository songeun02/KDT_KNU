# 시도별 전출입 인구수

## 대구 -> 타지역

```python
p_df=pd.read_excel('/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/data/시도별 전출입 인구수.xlsx')
p_df.drop(0,inplace=True)
p_df
```

```python
p_df['전출지별']= p_df['전출지별'].ffill()
p_df
```

```python
p_df_daegu=p_df[p_df['전출지별']=="대구광역시"].copy()
p_df_daegu
```

```python
p_df_daegu.info()
```

```python
p_df_daegu = p_df_daegu.replace('-',0)
p_df_daegu
```

```python
p_df_daegu.set_index(['전출지별','전입지별'],inplace=True)
p_df_daegu
```

```python
p_df_daegu.drop([('대구광역시','전국')],inplace=True)
p_df_daegu
```

```python
p_df_daegu.drop([('대구광역시','대구광역시')],inplace=True)
p_df_daegu
```

```python
p_df_daegu=p_df_daegu.astype(int)
p_df_daegu.info()
```

```python
p_df_daegu.max()
```

```python
p_df_daegu.idxmax()
```

```python
p_df_daegu.index.to_list()
```

```python
list_index=pd.DataFrame(p_df_daegu.index.to_list())
list_index
```

```python
list_index[1]
```

```python
len_list=len(p_df_daegu.index.to_list())
```

```python
plt.figure(figsize=[20,5])
plt.xticks(rotation=45)


for i in range(len_list):
  list_city=p_df_daegu.index.to_list()
  list_index_i=list_index[1]
  plt.plot(p_df_daegu.loc[list_city[i],:], label=list_index_i[i])
plt.legend()
plt.show()
```

## 전국

```python
p_df
```

```python
p_df_all=p_df[p_df['전출지별']=="전국"].copy()
p_df_all
```

```python
pd.set_option('future.no_silent_downcasting', True)
p_df_all = p_df_all.replace('-', 0)

p_df_all.drop(1,inplace=True)
p_df_all.reset_index(inplace=True)
p_df_all.drop('index',inplace=True,axis=1)
p_df_all
```

```python
p_df_all.set_index(['전출지별','전입지별'], inplace=True)
```

```python
p_df_all.info()
```

```python
p_df_all.max()
```

```python
p_df_all.idxmax()
```

```python
p_df_all.columns
```

```python
p_df_all_sub=pd.DataFrame()
```

```python
for i in range(1970,2017):
  p_df_all_sub[f'{i+1}-{i}']=abs(p_df_all[f'{i+1}']-p_df_all[f'{i}']) # 인원 이동 변동을 보는 것이므로 절대값 사용
```

```python
p_df_all_sub
```

```python
p_df_all_sub.info()
```

```python
p_df_all_sub.max()
```

```python
p_df_all_sub.idxmax()
```

```python
p_df_all_sub
```

```python
p_df_all_sub_index=pd.DataFrame(p_df_all_sub.index.to_list())
p_df_all_sub_index
```

```python
sub_list=p_df_all_sub.index.to_list()
```

```python
plt.figure(figsize=[20,5])

for i in range(len(sub_list)):
  list_city=p_df_all_sub.index.to_list()
  list_index_i=p_df_all_sub_index[1]
  plt.plot(p_df_all_sub.loc[list_city[i],:], label=list_index_i[i])
  plt.xticks(rotation=45)

plt.legend(loc='center right')
plt.show()
```
