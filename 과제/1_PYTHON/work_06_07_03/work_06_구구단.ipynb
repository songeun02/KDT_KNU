[1] for문 하나로 구구단 만들기

```python


for k in range(0, 81):
    i = k // 9 + 1  # 1부터 9까지의 곱하는 수 -> 2~10 몫
    j = k % 9 +2  # 2단부터 9단까지의 단 -> 2~10 나머지

    if j<=9:
        print(f'{j} x {i} = {j * i}', end='\t')
    else:
        print('')
```
