# 별찍기 다른 방법

## 직각삼각형 모양 별찍기

기본

```python
for i in range(5):
    for j in range(5):
      if j<=i:
          print('*', end= '')
    print()
```

--> 새로운 방법

```python
for i in range(5):
    for j in range(i+1):
        print('*', end= '')
    print()
```

## 대각선 모양 별찍기

```python

for i in range(5):
    for j in range(5):
        if i==j:
            print('*', end='')
        else:
            print(' ', end='')
    print()

```

--> 새로운 방법

```python
for i in range(5):
    for j in range(i+1):
        if i==j:
            print('*', end='')
        else:
            print(' ', end='')
    print()
```

```python
for i in range(5):
    for j in range(i+1):
        print('*' if i==j else ' ', end='\n' if i==j else '')
```
