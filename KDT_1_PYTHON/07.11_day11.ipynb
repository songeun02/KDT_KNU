## 모듈 내에서 일부 변수, 함수, 클래스만 포함하는 경우
- 형식 : from 모듈명 import 변수명/함수명.클래스명

```python
from math import pi
# math에서 pi만 가져옴
# -> 정의한 에 있는 것만 가져옴
```
