1. 서식지정자 방식 ( Format String )

```python
num1=2
num2=9

# print("%d + %d = %d" %num1 %num2 %(num1+num2))
# 위의 식이 아니라

print("%d + %d = %d" %(num1, num2, (num1+num2)))
```

```python
print("%d / %d = %f" %(num2, num1, (num2/num1)))

print("%s / %s = %s" %(num2, num1, (num2/num1)))
# 타입을 지정하기 어려우면 s로 작성한다.
# 컴퓨터에서 내부적으로 처리
# 정수를 문자로 형변환
```

2. F - string 방식

형식 : f'   {변수명 or 수식 or 데이터}   '

형식 : F'   {변수명 or 수식 or 데이터}   '

'{}'과 "{}" 는 동일

```python
print(f'{num1}+{num2}={num1+num2}')
print(f'{num2}/{num1}={num2/num1}')
```

## print() 매개변수
sep 매개변수 : 구분자 의미, 여러개 데이터를 보기 좋게 출력되도록 구분해주는 변수

end 매개변수 : 출력 데이터의 마지막 부분에 줄바꿈 문자를 추가해 놓은 변수




```python
 # 여러개 데이터 전잘 시 구분 문자 변경

print('010','1111','2222') # 기본 값 : 공백
print('010','1111','2222', sep='-')
print('010','1111','2222', sep='*')
print('010','1111','2222', '*') # 구분자를 위한 sep을 생략하면 안 됨
```

```python
# 20살
age = 20
print(age, '살')
print(age,'살',sep='') # '' : empty 문자 ( 비어있음 , 공백도 없음 )
```

```python
# 화면 출력 후에 문자 설정 -> [기본] 줄바꿈 '\n'

print(1)            # 1\n
print(2)            # 2\n
print(3)            # 3\n
print("\n")

print(1)            # 1\n
print(2, end=' ')   # 2_
print(3)            # 3\n
```

```python
# 출력 결과는 아래와 같고 print()함수는 4개 사용
# 1234567
# abcdefg ABCDEFG
# 1234567

print("1234567")
print("abcdefg", end=' ')
print("ABCDEFG")
print("1234567")
```

file 매개변수 : 데이터를 파일에 기록
'
- 파일 읽기 & 쓰기
  - 파일 열기 open()
  - 파일 읽기 또는 쓰기
  - 파일 닫기 close()

```python
FILENAME='result.txt'

# open(FILENAME) # 기본이 읽기로 열림

# 파일을 쓰기 모드(w)로 열기
f=open(FILENAME, mode='w') # 파일 정보를 f에 저장
f.write("home home home home")
f.close()

f=open(FILENAME, mode='w')
print("home home home home home", file=f) # 출력을 파일에 함
f.close()
```

# 실습 1

```python
# 입력 출력 실습

# 1. 데이터 저장 및 변수 생성
# 생년월일 , 띠, 혈액형


birthday_y = input("birthday year(0000년) : ")
birthday_m = input("birthday month(00월) : ")
birthday_d = input("birthday day(00일) : ")

chinese_zodiac = input("띠 : ")
blood_type= input("혈액형 : ")

print(f'나는 {birthday_y}년 {birthday_m}월 {birthday_d}일 {chinese_zodiac}띠 입니다')
print(f'혈액형은 노는 것을 좋아하는 {blood_type}형 입니다.')


# 2. 입력받은 데이터를 파일로 저장
# 좋아하는 계절, 좋아하는 나라, 여행가고 싶은 나라
# 파일명 자유

FILENAME='my_type.txt'

f=open(FILENAME, mode='w', encoding='utf-8') # 저장한 파일에서 한글이 깨질 경우 encoding='utf-8' 사용
like_weather = input("좋아하는 계절 : ")
like_country = input("좋아하는 나라 : ")
trip_country = input("여행가고 싶은 나라 : ")

print(f'나는 {like_weather}을 좋아하며 {like_country}를 좋아한다. 또한 {trip_country}에 여행가고 싶다 ', file=f)
f.close()

# f.write()는 줄바꿈이 불가능 하므로 f.write('\n')를 직접 입력해야 함
```

# 타입 캐스팅(Type Csting) / 형변환
자료형을 다른 종류의 자료형으로 변경

- 자동/묵시적 형변환 : 컴퓨터가 진행
  - 작은 것을 큰 것으로 변환 : 데이터 손실 없음
- 수동/명시적 형변환 : 개발자가 진행
  - 큰 것을 작은 것으로 변환

```python
a=10
b=3

print(a/b) # 컴퓨터가 자동적으로 형변환을 하여 소수점 형태로 출력

print("%s" %3.9) # 자동으로 s 를 f로 형변환하여 3.9 그대로 출력
```

# 객체 지향적 프로그래밍 ( OOP )

: 데이터가 객체 내에 캡슐화 되고 구성 요소 부분이 아닌 객체 자체가 운용되는 프로그래밍 접근 방식

- 추상화 : 생물/ 무생물에서 특징, 기능을 뽑아내는 작업 -> 클래스화

- 캡슐화 : 클래스가 가진 특징을 외부에 보이지 않도록 감춤, 감추는 것에 대한 공개 범위 존재

- 상속 : 클래스들의 공통된 특징과 기능을 추출하여 상위클래스 생성, 하위 클래스도 존재
  - 재사용을 하기 위함



```python
num1=-3.2
msg1='원더우먼'

msg2=''  # false
num2=0  # false


print(f'not {num1} = {not num1}')
print(f'not {num2} = {not num2}') # -> not false 이므로 true
print(f'not {msg1} = {not msg1}')
print(f'not {msg2} = {not msg2}') # -> not false 이므로 true
0
```
