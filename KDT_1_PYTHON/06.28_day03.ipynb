# list 데이터 자료형 살펴보기

- 여러 개의 데이터 저장하는 타입

- 다른 종류의 데이터도 함께 저장 가능

- 형식 : [data1, data2, ......]

```python
my=[16,170,70]

print(f'인덱싱 : {my[0]}, {my[1]}, {type(my[0])}')
print(f'슬라이싱 : {my[0:2]}, {type(my[0:2])}') # 괄호에 묶여져서 출력
```

=> 차이점 : 안에 들어있는 원소의 type이 그대로 적용되어서 다름

- 원소 / 요소 변경하기 -> 데이터 변경, 삭제

```python
datas = [1,3,5,7,9,11]

# 0번 원소 100으로 변경
datas[0] = 100
print(datas) # 성공적으로 변경 됨

del(datas[0])
print(datas) # 성공적으로 삭제 됨
```

```python
# 0번~2번 원소 까지 자리에 100으로 변경

datas[:3] = [100]*3
print(datas)

datas[:3] = [100]
print(datas)
```

빈 list에는 값이 없으므로 인덱스도 사용 못 하고 값도 불러올 수 없음

```python
nums=[10,20]
nums2=list([10,20])

print(f'nums의 id : {id(nums)}')
print(f'nums[0]의 id : {id(nums[0])}')
print(f'nums[1]의 id : {id(nums[1])}')
print("*"*40)
print(f'nums2의 id : {id(nums2)}')
print(f'nums2[0]의 id : {id(nums2[0])}')
print(f'nums2[1]의 id : {id(nums2[1])}')
```

# 튜플(Tuple) 데이터 자료형 살펴보기

리스트처럼 다른 자료형의 데이터들을 저장할 순 있지만 데이터를 수정, 삭제가 불가능한 read only이다.

또한 데이터 한 개가 들어가도 ',' 사용

- 형식 : (data1, ..... datan) or data1, ..... datan or (data, ) or data,

```python
datas=(1) # int
print(type(datas), datas, len(datas))
```

```python
datas=(1,) # tuple
print(type(datas), datas, len(datas))

datas=1, # tuple
print(type(datas), datas, len(datas))
```

## 튜플 데이터의 원소 / 요소 읽기

list와 동일하게 읽음

 변경만 불가능
  - 추가, 삭제, 수정 불가능

```python
datas=11,22,33,44,55

# 2번 요소 읽기
print(datas[2],datas[-3])

# 1,2,3 번 요소 읽기
print(datas[1:4])
```

```python
# 마지막 원소를 'A'로 변경
datas[-1] = 'A'
print(datas)
# 미지원으로 변경 불가능
```

```python
del(datas[-1])
print(datas)
# 미지원으로 삭제 불가능
```

변경이 불가능하도록 만들고 싶은 것을 튜플로 정의

형변환 가능

tuple -> list

- 수정가능하도록 만들기 위해서  

list -> tuple

- 수정 불가능하도록 만들기 위해서

```python
birthday=(2024,1,1)

# 1월 -> 3월 변경

birthday = list(birthday) # 원소가 여러개 있으므로 iterable하고 sequence 타입
print(birthday)
birthday[1]=3

# list를 다시 tuple로 형변환
birthday=tuple(birthday)
print(birthday, type(birthday))
```

# 내장함수 range()

- 숫자 범위를 생성하는 내장함수

- 형식

    - range(시작숫자, 끝숫자+1, 간격)

    - range(끝숫자+1) <- 0~ 끝숫자, 간격은 1



```python
nums = range(1,101)

print(f'nums값 : {nums} \n타입 : {type(nums)} \n개수 : {len(nums)}')
```

```python
# 원소 / 요소 읽기 => 인덱싱
print(nums[0], nums[-1])


# 원소 / 요소 여러개 읽기 => 슬라이싱
print(nums[0:10], nums[30:41]) # range를 잘랐으므로 range형태로 나옴
print(list(nums[0:10]), list(nums[30:41])) # list 형변환
print(tuple(nums[0:10]), tuple(nums[30:41])) # tuple 형변환
```
