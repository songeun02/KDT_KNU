# 함수 (Function) 이해 및 활용

```python
def multi3(num1,num2,num3):
    result = num1*num2*num3
    return result
    # return print(result) -> 함수의 결과값이 None
```

```python

# 사용자가 종료를 원할 때 종료 => 'x', 'X' 입력시 종료
# 연산 방식과 숫자 데이터 입력 받기
# 종료 메뉴 선택 시 프로그램 종료
# 반복 : 무한 반복


def add(num1,num2):
    return num1+num2

def sub(num1,num2):
    return num1-num2

def multi(num1,num2):
    return num1*num2

def div(num1,num2):
    if num2!=0:
        return num1/num2
    else:
        return 'num2가 0입니다'

def manu2():
    print(f'{"*":*^15}')
    print(f'{"계 산 기":^15}')
    print(f'{"*":*^15}')
    print(f'{"*  1. 덧  셈  *":15}')
    print(f'{"*  2. 뺄  셈  *":15}')
    print(f'{"*  3. 곱  셈  *":15}')
    print(f'{"*  4. 나눗셈  *":15}')
    print(f'{"*  5. 종  료  *":15}')
    print(f'{"*":*^15}')


def calc(func, num1,num2):
    num1=int(num1)
    num2=int(num2)
    print(f'결과 : {func(num1,num2)}')


while(1):
    manu2()

    choice=input("메뉴선택 : ")
    if choice.isdecimal(): # 입력받은 것이 정수가 맞으면 형변환
        choice=int(choice)
    else:
        print('0~9 사이 숫자만 입력하세요 ')
        continue


    if choice==5:
        print('계산기를 종료합니다')
        break
    elif choice==1:
        print('덧셈')
        num1,num2=input("숫자를 입력하시오 : ").split()
        calc(add,num1,num2)


    elif choice==2:
        print('뺄셈')
        num1,num2=input("숫자를 입력하시오 : ").split()
        calc(sub,num1,num2)
    elif choice==3:
        print('곱셈')
        num1,num2=input("숫자를 입력하시오 : ").split()
        calc(multi,num1,num2)
    elif choice==4:
        print('나눗셈')
        num1,num2=input("숫자를 입력하시오 : ").split()
        calc(div,num1,num2)
    else:
        print('선택한 메뉴는 없습니다. ')
```

# 람다표현식 또는 람다함수
- 1줄 함수, 익명 함수
- 형식 : lambda 매개변수 : 실행코드

## sorted()와 lambda

```python
names={1:'Kim',2:'Adam',3:'Zoo'}

# 정렬하기 -> 내장함수 sorted()

# key로 정렬
result = sorted(names) # sorted는 리스트 반환
print('오름차순 정렬 : ' ,result)

result = sorted(names)

# value로 정렬
result = sorted(names.items(), key=lambda x:x[1])
print('오름차순 정렬 : ' ,result)
```

```python
result=sorted("This is a test string from Andrew".split())
print(result)

result=sorted("This is a test string from Andrew".split(), key=str.lower)
print(result)
```

## map()과 lambda

```python
data=[11,22,33,44]

# 각 원소의 값에 곱하기 2해서 다시 저장

def multi2(value):
  return value*2

data2=list(map(multi2,data))
print(data2)
```

-> lambda 사용

```python
data2=list(map(lambda x:x*2, data)) # 힙 메모리 축소
print(data2)
```
