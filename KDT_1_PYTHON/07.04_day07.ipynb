```python
# [실습] 로또 프로그램을 만들어 주세요
# 1~45 범위 내에서 중복되지 않는 6개 추출

# list, set, dict, tuple을 사용하여 가능 but tuple은 수정 x , set은 인덱스로 접근 불가능하여 사용 x

import random as rad

cnt1=0
num=[0,0,0,0,0,0] # 빈 리스트 지정하지 않고 0으로 채움
# 리스트는 비어있으면 인덱스를 지정할 수 없음


while True:

    n = rad.randint(1,45)

    if n in num:
        continue
    else :
        num[cnt1]=n
        cnt1+=1

    if cnt1==6:
        break

print(num)
```

## 요소 순서 제어 메서드 reverse()
- 순서를 뒤집음
- 0번 -> -1번, -1번 -> 0번

```python
datas=[]

for _ in range(10):
  datas.append(rad.randint(1,30))

print(f'datas 개수 : {len(datas)}, {datas}')
```

rad.seed() : 동일한 랜덤 숫자 추출을 위한 기준점
- ()안에 숫자는 seed 테이블 번호

```python
rad.seed(10)

datas=[]

for _ in range(10):
  datas.append(rad.randint(1,30))

print(f'datas 개수 : {len(datas)}, {datas}')
```

```python
datas.reverse()
print(f'datas 개수 : {len(datas)}, {datas}')
```

- 메모리 사용량 감소, 속도 빠름

## 실습

```python
# [실습] a 리스트의 데이터를 b리스트에 담기
#       단, a 짝수값은 3을 곱하고 홀수값은 그대로해서 b리스트에 담기

a = [1,2,3,4,5,6]
b=[]

for num in a:
    if num%2==0:
        b.append(num*3)
    else:
        b.append(num)

print(f'a->{a}\nb->{b}')

# [1] 모든 원소를 새로운 리스트에 담기

c=[num for num in a]

print(f'a->{a}\nb->{b}\nc->{c}')
```

```python
# [2] 짝수 데이터만 새로운 리스트 c에 담기

# for num in a:
#   if not num%2: c.append(num*3)

# ->

c = [num*3 for num in a if not num%2]

print(f'a->{a}\nb->{b}\nc->{c}')
```

```python
# [3] 짝수 데이터는 3을 곱하고 홀수 데이터는 그대로 새로운 리스트 c에 담기

# for num in a:
#   if not num%2 : c.append(num*3)
#   else: c.append(num)

# ->

c = [num*3 if not num%2 else num for num in a]
# 값의 형태에 따라 num, num*3으로 달라지므로 if문을 앞으로 당겨야 함

# 내가 작성한 코드 : [num for num in a if num%2 else num*=3] -> 틀림

print(f'a->{a}\nb->{b}\nc->{c}')
```

```python
msg='Hello Good Luck'

for m in msg:
  print(m,msg.count(m))
```

알파벳이 중복되어서 나타나므로 총 알파벳의 개수를 알 수 없음

-> set() 사용하여 중복제거

```python
alpha=set(msg)
print(alpha)

for m in alpha:
  print(m,msg.count(m))
```

-> 딕셔너리로 만들기 위함

```python
alpha=set(msg)
save={}
print(alpha)

for m in alpha:
  print(m,msg.count(m))
  save[m]=msg.count(m)

print(save) # 정렬 필요하면 수행
```
