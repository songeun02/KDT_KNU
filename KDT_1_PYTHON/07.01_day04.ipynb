```python
data = {1,2,3,[1,2,3]}
print(f'data 타입 : {type(data)}, 원소 / 요소 개수 : {len(data)}개, 데이터 : {data}')
```

튜플은 가능

```python
data = {1,2,3,(1,2,3)}
print(f'data 타입 : {type(data)}, 원소 / 요소 개수 : {len(data)}개, 데이터 : {data}')
```

```python
data = {1,2,3,(1)}
print(f'data 타입 : {type(data)}, 원소 / 요소 개수 : {len(data)}개, 데이터 : {data}')
# (1)은 튜플이 아니고 1과 동일하므로 중복으로 처리

data = {1,2,3,(1,)}
print(f'data 타입 : {type(data)}, 원소 / 요소 개수 : {len(data)}개, 데이터 : {data}')
# 쉼표(,)를 찍으면 튜플이 되므로 중복으로 처리 x

data = {1,2,3,(1,)[0]} # 인덱스 0은 1 -> 중복 제거시 사용
print(f'data 타입 : {type(data)}, 원소 / 요소 개수 : {len(data)}개, 데이터 : {data}')
```

2. 생성자 메서드 ( 값을 넣음 )

## 내장함수 zip()

같은 인덱스의 요소끼리 묶어줌
- 같은 위치에 있는 것을 다 묶음
- 리스트 뿐만 아니라 튜플도 가능

```python
l1 = ['name','age','gender']
l2 = ['마징가',12,'남']
l3 = [False, True, True]
print(list(zip(l1,l2,l3)))

ds=dict(zip(l1,l2))
print(ds)
```

## 조건문 살펴보기

### 다중 조건문

- 조건이 2개 이상인 경우
- if 조건1 :

  ...........실행코드

  elif 조건2

  ...........실행코드

  elif 조건n

  ...........실행코드

  else:

  ...........실행코드

  => 조건이 True인 경우 실행코드 실행 후 나머지 조건은 검사 x

  => 조건이 False인 경우 아래 조건들 검사 실행

  -> 조건2에서 조건1과 겹치는 부분은 작성하지 않아도 ㅇ

### 중첩 조건문

- 조건문에 조건문이 존재하는 제어문

- if 조건식 :

  ...........실행코드

  ...........if 조건식 :

  .................실행코드

  .................실행코드
