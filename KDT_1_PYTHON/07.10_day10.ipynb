
# 지역변수 (Local Variable)

- 함수 내에 존재하며 함수에서만 사용 가능
- 프로그램 실행 시 메모리 존재
- 프로그램 종료 시 메모리에서 삭제

전역변수의 값을 변경할 경우 그냥 사용 x
- 단순히 읽기는 ㅇ

```python
# - 함수 기능 : 정수를 덧셈한 후 결과를 반환하는 함수
# - 함수 이름 : addInt
# - 매개 변수 : 0~N개 가변인자 -> *num => set은 중복인자 제거하므로 주로 tuple에 줌
# - 함수 결과 : 정수 result


def addInt(*num):
    total1=0
    for n in num:
        total1+=n
    return total1

def multiInt(*num):
    total1=1
    for n in num:
        total1*=n
    return total1+total


def multiInt2(*num):
    global total
    for n in num:
        total*=n
    return total


result1=addInt(1)
print(f'result1 : {result1}')


result2=multiInt(5)
print(f'result2 : {result2}')

print(f'전 : total = {total}')

result3=multiInt2(5)
print(f'result3 : {result3}')

print(f'후 : total = {total}')
```

전역변수


```python


persons=["Hong"] # list는 장바구니이므로 값이 아님, 그러므로 값은 그대로이지만 장바구니가 바뀌는 것으로 global 선언 안 해도 ㅇ
year=2024 # year는 값이 바뀌는 것이므로 global 선언

gender={'h':'남'}
# collection 타입은 global 선언 안 해도 ㅇ

# 사용자 정의 함수

def add_person(name):
    global year
    year+=1
    persons.append(name)
    gender[name]='여'

def remove_persons(name):
    persons.remove(name)
    gender.pop(name) # dict에서 삭제 방법


# 함수 호출

print(f'persons : {persons}')
print(f'year : {year}')
print(f'gender : {gender}')



add_person("Park")
print(f'persons : {persons}')
print(f'year : {year}')
print(f'gender : {gender}')


remove_persons("Park")
print(f'persons : {persons}')
print(f'year : {year}')
print(f'gender : {gender}')

```
