# 문자열 출력 형태

```python
# 데이터:빈칸채울문자 정렬 칸수

num=2

print(f'{num:0>3}입니다') # 오른쪽 정렬
print(f'{num:0^3}입니다') # 중앙 정렬
print(f'{num:0<3}입니다') # 왼쪽 정렬

# -> 빈 칸을 0으로 채움
```

```python
num=2

print(f'{num:=>10}입니다') # 오른쪽 정렬
print(f'{num:=^10}입니다') # 중앙 정렬
print(f'{num:=<10}입니다') # 왼쪽 정렬

# -> 빈 칸을 =으로 채움
```

# 제어문 - 반복문과 조건문 혼합

```python
# [실습]
# 메시지를 입력받는다
# 알파벳 대문자인 경우 소문자로, 소문자인 경우 대문자로
# 나머지는 그대로 되도록 출력하기

msg=input()

for i in msg:
    if 'a'<=i<='z':
        print(chr(ord(i)-32),end='')
    elif 'A'<=i<='Z':
        print(chr(ord(i)+32),end='')
    else:
        print(i,end='')
```

# break
반복을 중단 시키는 조건문과 함께 사용됨


```python
# [실습]
# 각 과목이 40점미만이면 과락이다. 4개 과목 평균이 60점 이상이면 합격

jumsu = [89,39,80,77]
isPass=True # 상황 체크 - flag 변수

# 과목별 40점 미만 체크
for jum in jumsu:
    if jum<40:
        print("당신은 과락")
        isPass=False
        break
    print('모든 과목 40점 이상')

# 합격 불합격 처리
if isPass:
    avg=sum(jumsu)/len(jumsu)
    if avg>=60:
        print(f'당신은 {avg}점으로 합격')
    else:
        print(f'당신은 {avg}점으로 불합격')

else:
    print(f'당신은 40미만인 과목으로 불합격')
```

## 제어문 - 반복문과 break

중첩 반목문일 경우의 break는 가장 가까이 있는 반복문만 종료

```python
#[실습] 단의 숫자만큼 구구단 출력

dan = int(input("출력 원하는 단 입력 : "))
isBreak=False # for에서 isBreak가 True면 종료되므로 False로 선언

for d in range(2,10):
    print(f'\n[{d}단]', end=' ')
    for n in range(1,10):
        print(f'{d} * {n} = {d*n:2}', end=' ')
        # 결과의 자리수를 맞춤. 오른정렬이 default , 왼쪽 정렬 하고싶으면 {d*n:<2}

        if n==d:
            isBreak=True
            break # 가장 근접한 if문만 종료 -> 외에 반복문도 종료하고 싶다면 새 변수 isBreak 사용

    if isBreak==True: # 가장 처음 for문을 종료하기 위함
        if d==dan:
            break

```

-> 내부 반복문 종료시 외부 반복문도 종료

--> 내부 반복문 종료여부를 변수 저장

---> 외부 반복문에서는 내부 반복문이 종료되면 함께 종료

```python
dan = int(input("출력 원하는 단 입력 : "))
isBreak=False # for에서 isBreak가 True면 종료되므로 False로 선언

for d in range(2,10):
    print(f'\n[{d}단]', end=' ')
    for n in range(1,10):
        print(f'{d} * {n} = {d*n:2}', end=' ')
        # 결과의 자리수를 맞춤. 오른정렬이 default , 왼쪽 정렬 하고싶으면 {d*n:<2}

        if n==d:
            isBreak=True
            break # 가장 근접한 if문만 종료 -> 외에 반복문도 종료하고 싶다면 새 변수 isBreak 사용

    if d==dan:
        break
```

# for vs while
- 원소를 읽을 땐 for를 더 주로 사용
- 명확한 수행 숫자가 안 정해지면 while 사용

```python
# [실습] 3단 출력하기, 단 while문 사용

# by me

num = list(range(1,10))
idx=0

while idx<len(num):

    print(f'3 * {num[idx]} = {3 * num[idx]}')
    idx+=1
```

```python
cnt=1
while cnt<10:
    print(f'3 * {cnt} = {3 * cnt}')
    cnt+=1
```
