### 문자열에 동일한 문자가 여러개 존재하는 경우

```python
msg='Good Luck Good'
```

```python
# 'o'의 인덱스 찾기 -> 첫번째 'o'인덱스

idx=msg.find('o') # find('o',0) : 뒤에 0은 생략됨
print(f'o 인덱스 :  {idx}')
```

```python
# 'o'의 인덱스 찾기 -> 두번째 'o'인덱스

idx=msg.find('o', idx+1) # 'od Luck Good' 부터 찾는다는 의미
print(f'o 인덱스 :  {idx}')
```

위의 코드를 여러번 사용하여 존재하는 'o'의 모든 인덱스 찾을 수 ㅇ

-> 반복문 사용

```python
msg='Good Luck Good'

cnt=msg.count('o')
idx_f=0

for i in range(cnt):
    idx_f=msg.find('o', idx_f+1)
    print(f'{i}번째 o 인덱스 :  {idx_f}')
```

## 문자열의 뒷부분부터 찾기하는 메서드 rfind(), rindex()

- rfind(문자,시작인덱스, 끝인덱스+1)
  - find(문자,시작인덱스,끝인덱스+1)
- rindex(문자,시작인덱스, 끝인덱스+1)
  - index(문자,시작인덱스, 끝인덱스+1)


-> 시작인덱스, 끝인덱스 사이에서 탐색

```python
msg='Happy'

# 'y' 인덱스 찾기
idx=msg.rfind('y')
print(f'y의 인덱스 : {idx}')

# 첫번째 'p' 인덱스 찾기
idx=msg.rfind('p')
print(f'첫번째 p의 인덱스 : {idx}')

# 두번째 'p' 인덱스 찾기
idx=msg.rfind('p',0,idx) # 끝 index는 포함 안 됨 -> 'Hap'에서 찾음
print(f'두번째 p의 인덱스 : {idx}')
```

```python
# # 사용자 정의 함수

# - 함수 기능 : 정수를 덧셈한 후 결과를 반환 하는 함수
# - 함수 이름 : add
# - 매개 변수 : 0~N개
# - 함수 결과 : 결과 result


def add(*nums):
    total=0
    for n in nums:
        total+=n
    return total

print(add())
print(add(1,1,1))
print(add(5,6))
print(add(8,9,10,11,22,3,4,4,555))
```

```python

def add(num1,num2):
    return num1+num2

def sub(num1,num2):
    return num1-num2

def multiple(num1,num2):
    return num1*num2

def divide(num1,num2):
    return num1/num2 if num2 else '0으로 나눌 수 없음'
```

```python
# [실습] 사용자로 부터 연산자, 숫자1, 숫자 2를 입력받아서 연산 결과 출력

num1,num2,tool=input("숫자1, 숫자2, 연산자 순서로 입력하시오 : ").split()
num1=int(num1)
num2=int(num2)


if tool=='+':
    print(add(num1,num2))
elif tool=='-':
    print(sub(num1,num2))
elif tool=='*':
    print(multiple(num1,num2))
elif tool=='/':
    print(divide(num1,num2))
else:
    print('올바르게 입력하시오')
```
