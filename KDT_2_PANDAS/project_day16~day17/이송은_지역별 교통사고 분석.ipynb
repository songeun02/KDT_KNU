### **수원시 권선구는 왜 교통사고가 많이 발생할까 ?**

수원시 권선구는 수원의 중심지로서 관광명소로 많은 장소들이 존재한다.
- 경기상상캠퍼스 : 문화예술과 자연이 어우러진 복합문화공간

- 마중공원 : 생태 연못, 초화원, 어린이 놀이터, 물놀이장, 반려견 쉼터, 저류지, 잔디광장 등의 시설 보유

- 올림픽공원 : 1987년 88서울올림픽 개최를 기념하기 위하여 조성

- 정수농장 : 딸기/토마토/감자 농가 체험

- 도나스데이 본점 : 수원 오목천동 도넛 맛집으로 유명한 도나스데이 본점

- 무궁화전박물관 :  한지를 이용하여 무궁화와 우리들의 들꽃을 실물과 같이 재현한 곳

- 포시즌힐링팜 : 체험형 치유농장

- 수달앤쥬 : 흔히 보기 힘든 동물들이 있는 실내 동물원


등이 있다.

=> 수원의 중심지이자 관광명소로 유명한 곳이 많아서 교통사고 발생 수가 가장 높다고 할 수 있다.

2018, 2019, 2020, 2021, 2022 년도의 교통사고 발생 수가

경기에서 가장 많고 경기도 내부에서는 수원시가 가장 많이 발생하였다는 사실을 알 수 있다.


또한 수원시 내에서는 수원의 중심지이자 관광명소로 유명한 곳이 많은 권선구에서

대부분의 교통사고가 발생하고 많은 사상자와 부상자가 발생한다는 것이 밝혀졌다.
