# DF 연결
- 연결
  - pandas.concat() 함수
  - DataFrame.append() 메서드

```python
import pandas as pd
```

```python
data1=[[11,22,33],[44,55,66]]
data2=[[20,21,22],[23,24,25]]
```

## concat()

```python
# 2차원 리스트를 DF로 저장

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
# [4] DF 연결
# [4-1] 함수 사용
# -[기본] 행 방향 연결
# -[설정] ignore_index=True

df12=pd.concat([df1,df2], ignore_index=True) # 기존의 인덱스를 무시하고 새로 넣어서 정렬
df12
```

```python
# -[기본] 열 방향 연결
# -[설정] ignore_index=True, axis=1

df12=pd.concat([df1,df2], axis=1, ignore_index=True) # 기존의 인덱스를 무시하고 새로 넣어서 정렬
df12
```

```python
df2
```

```python
# [5] DF 연결 - 컬럼 수/ 컬럼 이름이 다른 경우

df3=pd.DataFrame(data2, index=[0,2],columns=[0,2,4])
df3
```

```python
df1
```

```python
df3
```

### 행방향 연결

```python
# [기본] 행방향 연결 axis=0, join="outer"

df13=pd.concat([df1,df3])
df13
```

```python
df13=pd.concat([df1,df3], ignore_index=True)
df13
```

```python

df13=pd.concat([df1,df3], join="outer")
df13
```

```python
# [설정] 행방향 연결 axis=0, join="inner"

df13=pd.concat([df1,df3], join="inner")
df13
```

### 열 방향 연결

```python
# [기본] join="outer"
# [설정] 열방향 연결 axis=1

df13=pd.concat([df1,df3], axis=1)
df13
```

```python
# [기본] join="outer"
# [설정] 열방향 연결 axis=1

df13=pd.concat([df1,df3], axis=1,join='inner')
df13
```

# DF 통합
- DF 에서 특정 컬럼을 기준으로 통합해주는 기능
- 함수 : pandas.merge()
  - default : inner
    - 가장 많이 쓰는 것을 default로 제공
  - 기준 지정 안 하면 동일한 열을 기준으로 동작

- 메서드 : DF.join()


## merge() 함수

```python
# xlsx 파일 -> DF로 저장
file_name1='/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/data/07_17/stock price.xlsx'
file_name2='/content/drive/MyDrive/KDT AI, 빅데이터 전문가 양성과정/data/07_17/stock valuation.xlsx'

price_df=pd.read_excel(file_name1)
val_df=pd.read_excel(file_name2)
```

```python
print(price_df.head(3), val_df.head(3), sep='\n\n')
```

```python
# [4] DF 합치기 -> pandas.merge() 함수

stock_df=pd.merge(price_df, val_df) # id를 기준으로 같은 것만 들어옴
stock_df.info()
```

```python
stock_df
```

```python
# DF 별 특정 컬럼 지정

stock_df=pd.merge(price_df, val_df,left_on='stock_name',right_on='name') # 이름을 기준으로 합침 -> 동일
stock_df
```

컬럼명은 다르지만 데이터는 동일하면 left_on, right_on 사용

```python

stock_df=pd.merge(price_df, val_df,left_on='stock_name',right_on='bps')  # 안 맞아서 불가능
stock_df
```

```python
stock_df=pd.merge(price_df, val_df,how='outer') # 모든 데이터 결합
stock_df
```

```python
stock_df=pd.merge(price_df, val_df,how='right') # 오른쪽 데이터를 다 살리고 그에 따라 왼쪽이 없으면  NaN
```

```python
stock_df=pd.merge(price_df, val_df,how='cross') # 하나하나 다 대응시키므로 13x10으로 130개 나옴 - 데이터 부족 시 강제로 데이터 늘리기 위해 사용
stock_df
```
