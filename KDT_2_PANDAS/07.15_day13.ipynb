## 행/열 인덱스 변경 방법

### (1) 메서드 사용
- rename() : 부분 일부 변경


### (2) 속성 사용
- index = [새로운 인덱스]
- columns = [새로운 컬럼]

## 행/열 삭제 메서드
- drop()
  - 행 삭제
    - axis=0, 삭제하고 싶은 인덱스 지정
    - index = 삭제하고 싶은 인덱스 지정 -> axis 와 상관 x
  - 열 삭제
    - axis=1, 삭제하고 싶은 인덱스 지정
    - columns = 삭제하고 싶은 인덱스 지정 -> axis 와 상관 x
    


## inplace : default = False
- 원본 삭제하고 싶으면 True

## 행/열 추가 메서드
- 행 추가 => loc[새로운 행] = 새로운 값
  - 새로운 값 1개 : 전부 동일한 값
  - 새로운 값의 수 = 열의 수 만큼 값
- 열 추가 => [새로운 열] = 새로운 값
  - 새로운 값 1개 : 전부 동일한 값
  - 새로운 값의 수 = 열의 수 만큼 값

## 원소 선택/추가 메서드
- 열 선택 ==> 변수명[열인덱스]
- 행 선택 ==> 변수명.loc[행인덱스] / 변수명.iloc[행인덱스]
- 원소 선택 ==>
  - 변수명.loc[행인덱스,열인덱스] / 변수명.iloc[행인덱스,열인덱스]
  - 변수명.loc[행인덱스][열인덱스] / 변수명.iloc[행인덱스][열인덱스]

```python
jumsu={'name':['마징가','태권브이','원더우먼'], 'kor':[97,49,66],'eng':[76,87,98], 'math':[78,97,75]}
```
