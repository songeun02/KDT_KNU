# 클래스(class)

- 객체지향언어(OOP)에서 데이터를 정의하는 자료형
- 데이터를 정의할 수 있는 데이터가 가진 속성과 기능 명시
- 구성요소 : 속성/attribute/field + 기능/method

```python
# 클래스 정의 : 햄버거를 나타내는 클래스
# 클래스 이름 : Burger
# 클래스 속성 : 번, 패티, 야채, 치즈
# 클래스 기능 : 햄버거 설명 기능

class Burger:

    # 힙 영역에 객체 생성 시 속성값 저장 기능
    def __init__(self,bread, patty,veg, kind): # 매개변수로 아래에서 정의한 값들 받음
        self.bread=bread
        self.patty=patty
        self.veg=veg
        self.kind=kind
        # __는 이미 선언되어 있는 함수
        # __init__ : 초기화
        # self : 클래스 안에 정의된 함수는 클래스 내부 개체만 사용가능하므로 자동으로 찾을 수 있도록 해줌

    # 클래스 기능 즉 메서드 => 매개변수에 self 넣어야 함
    def printInfo(self):
        print(f'브랜드 종류 : {self.kind}')
        print(f'빵 종류 : {self.bread}')
        print(f'패티 종류 : {self.patty}')
        print(f'야채 종류 : {self.veg}')

     # 속성을 변경하거나 읽어오는 메서드 -> getter/setter 메서드 - 객체지향

    def get_bread(self):
      return self.bread   # 원래 있던 인스턴스에 속성값 반환

    def set_bread(self,bread):
      self.bread=bread

# 객체 생성

# 불고기 버거 객체생성
burger1=Burger('브리오슈','불고기','양상추 양파 토마토', '롯데리아')
# 매개변수 5개 중 self는 파이썬에서 메모리 주소를 자동으로 넣어줌

# 치즈 버거 객체생성
burger2=Burger('참깨곡물빵','쇠고기','치즈 양상추 양파 토마토','맥도날드')

# 버거 정보 확인
burger1.printInfo() # 직접접근

# 속성 읽는 방법 : 1) 직접 접근 읽기 2) 간접 접근 읽기 - getter 메서드 사용
print(burger1.bread, burger1.get_bread()) # 간접접근

# 속성 변경 방법 : 1) 직접 접근 변경 2) 간접 접근 변경 - setter 메서드 사용
burger1.bread='들깨빵' # 직접접근
burger1.set_bread('올리브빵')
print()
burger2.printInfo()

# 특정 개체로 만들어진 것 : instance => 속성만 받음
# 클래스는 instance의 주소를 저장

print()
# 버거 인스턴스 속성 변경
burger1.patty='치킨'
burger1.printInfo()


```

getter, setter 메서드 추가

```python
# 클래스 정의 : 햄버거를 나타내는 클래스
# 클래스 이름 : Burger
# 클래스 속성 : 번, 패티, 야채, 치즈
# 클래스 기능 : 햄버거 설명 기능

class Burger0:

    kind="맥도날드" # 공통적으로 사용하는 것은 외부로 뺌 ( 모두 사용 ) => 클래스 속성

    def __init__(self,bread, patty,veg):
        # __init__ 내부 : 인스턴스 속성
        self.bread=bread
        self.patty=patty
        self.veg=veg

    def printInfo(self):
        print(f'브랜드 종류 : {self.kind}')
        print(f'빵 종류 : {self.bread}')
        print(f'패티 종류 : {self.patty}')
        print(f'야채 종류 : {self.veg}')


burger1=Burger0('브리오슈','불고기','양상추 양파 토마토')

burger2=Burger0('참깨곡물빵','쇠고기','치즈 양상추 양파 토마토')

burger1.printInfo()
print()
burger2.printInfo()

```

## 리스트로 DF 구성

```python
import pandas as pd
```

```python
# 데이터 준비
jumsu=[[98,99,100],[67,81,80]]
```

```python
# 데이터 프레임으로 데이터 저장
jumsuDF=pd.DataFrame(jumsu)
```

```python
# 데이터 프레임으로 저장된 데이터 확인
print(jumsuDF)
```

```python
# 데이터 프레임 속성 읽기

print(jumsuDF.index , '\n')
print(jumsuDF.columns, '\n')
print(jumsuDF.values, '\n')
print(jumsuDF.dtypes) # 데이터 프레임은 dtypes - 여러개이므로 -> column 별로 타입 알려줌
print(type(jumsuDF.dtypes), '\n')
print(jumsuDF.ndim, '\n')
print(jumsuDF.shape, '\n') # 몇 행 몇 열 인지 알려줌 -> 튜플로 나옴
```

```python
# DF 안에 있는 모든 원소를 하나씩 출력

for i in jumsuDF.index:
  for j in jumsuDF.columns:
    print(jumsuDF[i][j])
```

## DF 행/열 인덱스 변경
- 방법 1 : 속셩명으로 변경 -> 전체 변경
- 방법 2 : 메서드로 변경 -> 일부 변경 = rename()

```python
data={'name':['홍길동','마징가','원더우먼'],
      'job':['의적','로봇','영웅'],
      'gender':['남','남','여'],
      'age':[100,200,300],
      'height':[180,300,200]}
```

```python
dataDF=pd.DataFrame(data)
dataDF
```

```python
# [실습] 열 인덱스를 이름, 직업, 성별,나이, 키로 변경
dataDF.columns=['이름','직업','성별','나이','키']
dataDF
```

##### 열 인덱스 일부 변경

```python
# 키 -> 신장으로 변경
dataDF.rename(columns={'키':'신장'})
```

메서드는 원본 변경 x , 복사본에 변경

원본 바꾸고 싶다면 inplace=True로 주면 원본에 변경
- 복사본 존재 x

```python
dataDF.rename(columns={'키':'신장'}, inplace=True) # return이 None
dataDF
```

##### 행 인덱스 일부 변경

```python
# [실습] 행 인덱스 0,1,2 -> P0,P1,P2로 변경

dataDF.index=['P0','P1','P2']
dataDF
```

```python
# P2 -> END 변경

dataDF.rename(index={'P2':'END'})
```

```python
dataDF.rename(index={'P2':'END'},inplace=True)
dataDF
```

## DF에서 행/열 추가

- 행 방향 추가
- 열 방향 추가

### 행 추가

- loc 만 가능
  - iloc[]는 인덱스가 있어야 가능한데 없으므로 불가
- 변수명.loc['추가할 행이름'] = 데이터**
- 데이터 수 == 컬럼수 경우 : 행추가 가능
- 데이터 수 == 1개 경우 : 행추가 가능
- 그 외의 경우 행 추가 불가

```python
# 1개 행 추가
dataDF.loc[3]=['베트맨','박쥐','남',40,190]

# iloc가 안 됨 -> iloc는 시스템에서 제공하는 것이므로 안 되지만 loc로 사용자가 지정하는건 됨

dataDF
```

```python
dataDF.iloc[4]=['베트맨','박쥐','남',40,190]
# iloc 불가능
```

```python
# 1개 행 추가 : 데이터 수를 1개로 -> 모든 열에 동일한 값 저장

dataDF.loc[5]=100
dataDF
```

```python
# 1개 행 추가 : 데이터 수를 2개로 -> error -> 개수를 맞춰서 주거나 아니면 1개만 줘야 함

dataDF.loc[6]=['스파이더맨','거미']
dataDF
```

### 열 추가
- 변수명.loc['추가할 열이름'] = 데이터
- 데이터 수 == 행 수 경우 : 열 추가 가능
- 데이터 수 == 1개 경우 : 열 추가 가능
- 그 외의 경우 열 추가 불가

```python
dataDF
```

```python
# 열 추가 -> weight 추가 데이터 수 == 행 수
dataDF['weight']=[90,100,80,78,68,90]
dataDF
```
