# Series

## Series 데이터 자료형
- 판다스에서 1줄의 데이터를 저장하는 자료형/데이터 타입
- 형식 : pandas.Series()
- 구성 : Index+Data
  - index는 시스템이 제공
  - dtype은 데이터 타입

```python
 # 모듈 로딩
 # import pandas as pd
```

Series()도 생성자 함수

```python
# 시리즈 데이터 생성

sr0=pd.Series(10)
print(sr0)
print()

sr1=pd.Series([10,20,30]) # 묶여서 들어가야 하므로 리스트로 넣음
print(sr1)
print()

sr1=pd.Series([10,20.,30])  # 하나만 데이터 형태 바꿔도 다 변경
print(sr1)
print()

sr1=pd.Series([10,20.,'30'])  # 다 섞여있음(정수,실수,문자) -> object
print(sr1)
```

```python
# 데이터 준비
data1=['A','B','C']
data2={'name':'Hong','age':10}
```

-> 딕셔너리는 키가 인덱스로 들어감

```python
# [리스트 데이터] 생성된 시리즈 객체 속성 확인

print("index : ", sr1.index)
print("values : ", sr1.values)
print("dtype : ", sr1.dtype)
print("ndim : ", sr1.ndim)
```

```python
# [딕셔너리 데이터] 생성된 시리즈 객체 속성 확인

print("index : ", sr2.index)
print("values : ", sr2.values)
print("dtype : ", sr2.dtype)
print("ndim : ", sr2.ndim)
```

#### 시리즈 객체의 원소/요소 읽기

: 변수명[인덱스]


```python
# sr1 객체의 원소 읽기
print(sr1[0],sr1[1],sr1[2])
```

```python
print(sr1[-1])
```

=> 음수 인덱스는 error 발생

==> 위에서 RangeIndex(start=0, stop=3, step=1)라고 범위 정의됨

```python
# sr2 객체의 원소 읽기
print(sr2['name'],sr2['age'])
```

```python
# sr1객체에서 2개의 원소 읽기 ㅣ
# - 0번 원소와 2번 원소 읽기

print(sr1[0], sr1[1]) # 한 개씩 따로따로 가져옴
print()

print(type(sr1[0]), type(sr1[1]))
print()

print(sr1[[0,2]])     # 두 개 한꺼번에 -> series로 가져옴
print(type(sr1[[0,2]]))
```

```python
print(sr1[0:2]) # 슬라이싱도 가능
print(type(sr1[0:2])) # 여러개이므로 series
print()

print(sr1[0:2:2])
print()

print(sr1[:2])
print()

print(sr1[:]) # 처음부터 끝까지
```

==> 슬라이싱

===> 정수 인덱스 : [시작:끝+1]

```python
print(sr2[:'age'])
print()

print(sr2[:'name'])
```

#### 인덱스 & 인덱스레이블(네임)
