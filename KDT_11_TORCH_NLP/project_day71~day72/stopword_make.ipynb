```python
with open('stopwords.txt', 'r', encoding='utf-8') as f:
    stopwords_2 = f.readlines()
stopwords_2
```

```python
len(stopwords_1), len(stopwords_2)
```

```python
len(stopwords_1 + stopwords_2)
```

```python
stopwords_list = stopwords_1 + stopwords_2
```

```python
stopwords_list_pr = []

for data in stopwords_list:
    data = data.replace('\n','').strip()
    stopwords_list_pr.append(data)
```

```python
stopwords_list_pr
```

```python
stopwords_set = set(stopwords_list_pr)
```

```python
len(stopwords_set)
```

```python
file_name = 'stop_words.txt'

with open(file_name, 'a') as file:
    file.write('\n'.join(stopwords_set))  # '\n' 대신 ', '를 사용하면 줄바꿈이 아닌 ', '를 기준으로 문자열 구분함
```
