```python
def set_customFont(font_path):
    # 한글  폰트 설정 -> 폰트 매니저 모듈 사용
    from matplotlib import font_manager as fm
    from matplotlib import rc

    # 폰트 패밀리 이름 가져오기
    font_name=fm.FontProperties(fname=font_path).get_name() # FontProperties : 폰트 속성을 가져옴 -> 경로 주고 이름 받아옴

    # 새로운 폰트 패밀리 이름 지정

    rc('font',family=font_name) # 지금 현재 사용하는 폰트를 새로 바꿈

set_customFont(r'C:\Windows\Fonts\Hancom Gothic Regular.ttf')
```

```python
def draw_line_graph1(title,columns,labels,xlabel,ylabel):
    for col in columns:
        plt.plot(df.loc[col],'bo--',label=f'{col}') # 들어가는 데이터가 바뀌므로 매개변수로 선언

    plt.title(f'----[{title}]----')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(labels)
    plt.show()
```

```python
def draw_line_graph1_1(title,columns,labels,xlabel,ylabel,fmfs):
    for idx,col in enumerate(columns):
        plt.plot(df.loc[col],fmfs[idx])

    plt.title(f'----[{title}]----')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(labels)
    plt.show()
```

```python
def draw_line_graph2(title,columns,labels,xlabel,ylabel):
    for col in columns:
        plt.plot(df.loc[col],'bo-',label=f'{col}')

    for idx, col in enumerate(columns):
        plt.plot(df.loc[col],'bo-',label=f'{labels[idx]}')

    plt.title(f'----[{title}]----')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

```python
# 하나의 figure에서 수력/화력/원자력/합계 그래프 넣기 - 1

fig=plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

# - 각 subplot에 그래프 넣기
ax1.plot(df.loc[('남한','수력')], 'bo-',label='남한-수력')
ax2.plot(df.loc[('남한','화력')], 'bo-',label='남한-화력')
ax3.plot(df.loc[('남한','원자력')], 'bo-',label='남한-원자력')
ax4.plot(df.loc[('남한','합계')], 'bo-',label='남한-합계')

# 각 subplot 별로 설정
ax1.set_title("남한 수력 발전")
ax1.set_xlabel('년도')

ax2.set_title("남한 화력 발전")
ax2.set_xlabel('년도')

ax3.set_title("남한 원자력 발전")
ax3.set_xlabel('년도')

ax4.set_title("남한 에너지 발전")
ax4.set_xlabel('년도')

# 화면에 띄우기
plt.tight_layout()
plt.legend()
plt.show()
```

# Matplotlib 활용 시각화 - 막대 그래프
- 가장 많이 사용되는 그래프
- 데이터를 비교할 때 사용

```python
# [2] 데이터

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']
```

```python
# [3] 데이터 시각화

plt.bar(fruits,counts)
plt.show()
```

```python

plt.bar(fruits,counts, width=0.3) # 두께 조절
plt.show()
```

```python
plt.bar(fruits,counts, color=bar_colors) # 막대 별로 색깔 다르게
plt.show()
```

```python
plt.bar(fruits,counts, label=bar_labels)
plt.legend()
plt.show()
```

수평 막대 그래프

```python
# [3] 데이터 시각화 - 수평막대그래프
# bar -> barh

plt.barh(fruits,counts, height=0.3, color=bar_colors,label=bar_labels)
plt.legend()
plt.show()
```

수직 막대 그래프 + 선 그래프

```python
# [3] 데이터 시각화 - 수직막대그래프
# bar -> barh

plt.bar(fruits,counts, width=0.3, color=bar_colors,label=bar_labels)
plt.plot(counts,'bo--')
plt.legend()
plt.show()
```
