```python
import pandas as pd
import df_util as util
```

```python
file1='test_data.csv'
file2='test_data2.csv'
```

```python
df1=pd.read_csv(file1)
df2=pd.read_csv(file2)
```

```python
util.checkDataFrame(df1, 'df1')
util.checkDataFrame(df2, 'df2')
```

```python
# [5] 행 인덱스 설정
# [5-1] df1의 첫번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])
df11=df1.set_index(['index'])
df11.index.name=None
print(df11.index)

# 'A'행 데이터 추출
df11.loc['A']
```

```python
df11
```

```python
# 'A', 'C'행 추출
df11.loc[['A','C']]
```

```python
# [5-2] df1의 첫번째,두번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])
df12=df1.set_index(['index', 'col1'])
print(df12.index)

# ('A', 8) 행 추출
df12.loc[('A', 8)]
```

```python
# ('A', 8), ('A', 1) 행 추출
df12.loc[ [('A', 8), ('A', 1)] ]
```

```python
# 멀티인덱스의 속성들 확인
mIdx=df12.index
print(f'nlevels : {mIdx.nlevels}')
print(f'levels  : {mIdx.levels}')
print(f'names   : {mIdx.names}')

# mIdx.names=[None,None]
# print(f'names   : {mIdx.names}')
# mIdx.names[0]='A'
mIdx.set_names('a', level=0)
```

```python
# 멀티인덱스에서 level0이 동일한 행 추출
df12.loc['A']
```

```python
# 멀티인덱스에서 level1이 동일한 행 추출
#df12.loc[8]    # Error 발생
df12.xs(8, level=1)
```

```python
df12
```

```python
#  멀티인덱스에서 행 추가
# => 인덱스에 행 추가 => 변수명.loc[행인덱스]=값
df12.loc[('D', 9), :] = [9,9]
```

```python
# df12.loc[('D',9),:] = pd.Series({'col2':10,'col3':10},index=['D',10])
# -> 오류 발생
```

```python
sr1 = pd.Series({'col2':10,'col3':10})
sr1.index = ['D',10]
sr1.T
```

```python
df12.loc[('AA',102),:] = [sr1]

# 시리즈가 iterable 데이터가 아니라서 리스트로 만들어 이터러블로 바꾸기
# value Error 에서 iterable 이라는 말 나오면 일단 이터러블 타입 으로 변환
```
