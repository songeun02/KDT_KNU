match 메소드

```python
import	re

tel_checker = re.compile(r'^(\d{2,3})-(\d{3,4})-(\d{4})$')

print(tel_checker.match('02-123-4567'))

match_groups =	tel_checker.match('02-123-4567').groups() # 튜플로 출력
match_group =	tel_checker.match('02-123-4567').group() # 하나로 출력

print(match_groups)
print(match_group)

print(tel_checker.match('053-950-45678'))
print(tel_checker.match('053950-4567'))
```

```python
tel_number  = '053-950-4567'
tel_number = tel_number.replace('-','')
print(tel_number)

tel_checker1 = re.compile(r'^(\d{2,3})(\d{3,4})(\d{4})$')
print(tel_checker1.match(tel_number))
print(tel_checker1.match('0239501234'))
```

```python
tel_checker = re.compile('^(\d{2,3})-(\d{3,4})-(\d{4})$')
m = tel_checker.match('02-123-4567')

print(m.groups())
# group(index) : 해당 인덱스에 매칭된 문자열 반환 
print('group() : ', m.group())
print('group(0) : ',m.group(0))
print('group(1) :', m.group(1))
print('group(2,3) :', m.group(2,3))
print('start() : ', m.start()) # 매칭된 문자열의 시작 인덱스
print('end() : ',m.end()) # 매칭된 문자열의 마지막 인덱스 + 1
```

```python
tel_checker = re.compile('^(\d{2,3}-)(\d{3,4}-)(\d{4})$')
m = tel_checker.match('02-123-4567')

print(m.groups())
print('group() : ', m.group())
print('group(0) : ',m.group(0))
print('group(1) :', m.group(1))
print('group(2,3) :', m.group(2,3))
print('start() : ', m.start()) # 매칭된 문자열의 시작 인덱스
print('end() : ',m.end()) # 매칭된 문자열의 마지막 인덱스 + 1
```

전방 탐색

```python
lookahead1 = re.search('.+(?=won)','1000 won')
if (lookahead1 != None):
  print(lookahead1.group())
else:
  print('None')

lookahead2 = re.search('.+(?=am)','2023-01-26 am 10:00:01')
print(lookahead2.group())

lookahead3 = re.search('\d{4}(?!-)','010-1234-5678')
print(lookahead3)
```

```python
lookbehind1 = re.search('(?<=am).+','2023-01-26 am 11:10:01')
print(lookbehind1)

lookbehind2 = re.search('(?<=:).+', 'USD: $51')
print(lookbehind2)

lookbehind4 = re.search(r'\b(?<!\$)\d+\b','I paid $30 for 100 apples.')
print(lookbehind4)
```

링크간 무작위 이동하기

```python
import random
import re

random.seed(None) # 난수 발생기 초기화 

def getLinks(article_url):
  html = urlopen('https://en.wikipedia.org'	+	article_url)
  bs = BeautifulSoup(html, 'html.parser')
  bodyContent = bs.find('div',{'id':'bodyContent'})
  wiki_url = bodyContent.find_all('a',href=re.compile('^(/wiki/)((?!:).)*$'))
  return wiki_url

links = getLinks('/wiki/Kevin_Bacon')
print('link 길이 : ', len(links))
while len(links)>0:
  new_article = links[random.randint(0,len(links)-1)].attrs['href']
  # 발견한 링크 중 무작위 선택 
  print(new_article)
  links = getLinks(new_article)
```

전체 사이트 데이터 수집 소스

```python
pages = set() # 동일한 페이지를 두 번 크롤링 하는 것 방지 
count = 0
def getLinks(page_url):
  global pages
  global count
  html = urlopen('https://en.wikipedia.org{}'.format(page_url))
  bs = BeautifulSoup(html, 'html.parser')
  try:
    print(bs.h1.get_text())
    print(bs.find('div',attrs={'id':'mw-content-text'}).find('p').text)
  except AttributeError as e:
    print('this page is missing something! Continuing: ',e)
    pattern = '^(/wiki/)((?!:).)*$'
    for link in bs.find_all('a',href= re.compile(pattern)):
      if 'href' in link.attrs:
        if link.attrs['href'] not in pages:
          new_page = link.attrs['href']
          print('-'*40)
          count += 1
          print(f'[{count}] : {new_page}')
          pages.add(new_page)
          getLinks(new_page)

getLinks('')
```
