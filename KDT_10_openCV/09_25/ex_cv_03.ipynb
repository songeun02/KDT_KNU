# 이미지 Drawing
- 이미지 객체 ndarray 준비 필요

```python
# 모듈 로딩 
import cv2
import numpy as np
```

```python
# 데이터 관련 변수 설정 
DIR_PATH = '../09_25/Image/'
```

```python
# 이미지 객체 
# - 도화지가 들어갈 데이터 생성 
bg_data = np.full((500,500),255) # 255로 채움 - white

# 이미지 파일 생성 
BG_FILE_W = DIR_PATH + 'white_bg.jpg'
cv2.imwrite(BG_FILE_W, bg_data)

print(bg_data)
```

```python
# 이미지 객체
# - 도화지가 들어갈 데이터 생성 
bg_data = np.full((500,500),0) # 0으로 채움 - black

# 이미지 파일 생성 
BG_FILE_B = DIR_PATH + 'black_bg.jpg'
cv2.imwrite(BG_FILE_B, bg_data)

print(bg_data)
```

```python
# 이미지 객체 - 3차원
# - 도화지가 들어갈 데이터 생성 
bg_data = np.full((500,500,3),255) # 255로 채움 - white

# 이미지 파일 생성 
BG_FILE_W_3 = DIR_PATH + 'white_bg_3.jpg'
cv2.imwrite(BG_FILE_W_3, bg_data)

print(bg_data)
```

```python
# 도화지 이미지 가져오기

white_bg = cv2.imread(BG_FILE_W_3)
white_bg.shape, white_bg.ndim
```

새 도화지에 그릴려면 위의 코드 재실행 후 그리기

- 선 그리기

```python
# 선그리기 

cv2.line(white_bg, (50,50), (250,250), (255,0,0),5)
cv2.line(white_bg, (100,50),(250,250),(0,255,0),5)
cv2.line(white_bg, (250, 50), (250,300), (255, 0, 120), 10)
cv2.line(white_bg, (-10, 50), (250,450), (255, 0, 255), 20)
# 도화지 이미지, 시작점, 끝 점, 선 색깔, 선 두께 

cv2.imshow('DRAWING', white_bg) # 창 이름, 도화지 이미지
cv2.waitKey()
cv2.destroyAllWindows()
```

- 사각형 그리기

```python
# 사각형 그리기 

cv2.rectangle(white_bg, (50,50), (120,70), (0,255,255), 4)
# 도화지 이미지, 시작점, 끝 점, 선 색깔, 선 두께 

cv2.imshow('DRAWING', white_bg) # 창 이름, 도화지 이미지
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 사각형 그리기 

cv2.rectangle(white_bg, (50,50), (120,70), (0,255,255), -1)
# 도화지 이미지, 시작점, 끝 점, 선 색깔, 선 두께
# 선 두께 = -1 : 사각형 안 채움  

cv2.imshow('DRAWING', white_bg) # 창 이름, 도화지 이미지
cv2.waitKey()
cv2.destroyAllWindows()
```

- 다각형 그리기
    - numpy 배열로 좌표 생성

```python
# 다각형 그리기 

cv2.polylines(white_bg, [np.array([[50,50],[150,150],[100,140],[200,240]], dtype = np.int32)], False, (255,0,0)) # 번개모양 선 좌표 
cv2.polylines(white_bg, [np.array([[350,50],[250,200],[450,200]], dtype = np.int32)], False, (0,0,0), 10) # 삼각형 좌표 
cv2.polylines(white_bg, [np.array([[150,300],[50,450],[250,450]], dtype = np.int32)], True, (0,0,255), 10) # 삼각형 좌표 
cv2.polylines(white_bg, [np.array([[350,250],[450,350],[400,450],[300,450],[250,350]])],True, (0,0,0))

cv2.imshow('DRAWING', white_bg) # 창 이름, 도화지 이미지
cv2.waitKey()
cv2.destroyAllWindows()

```
